Setting up the Node Class

```python
class QAOANode:

    neighbors = []

    def __init__(self, point, name):
        self.point = point
        self.coordinates = (point["geometry"].x, point["geometry"].y)
        self.name = name

    def changeName(self, new_name):
        self.name = new_name
        
    def __str__(self):
        return self.name
```

```python
coal_nodes = []
coal_coords = []

powerplant_nodes = []
powerplant_coords = []

city_nodes = []
city_coords = []

#the row is what gives the point data itself -- state, coordinates, etc


for idx, row in coal_file.iterrows():
    
    if idx ==  9: break
    node = QAOANode(row, "coal_node:" + str(idx))

    coal_nodes.append(node)
    coal_coords.append(node.coordinates)
    
    
    

for idx, row in cities_file.iterrows():
    
    node = None
    
    if idx == 9: break
    
    if idx < 4:
        node = QAOANode(row, "power_node:" + str(idx))
    else:
        node = QAOANode(row, "city_node:" + str(idx))
    
    
    city_nodes.append(node)
    city_coords.append(node.coordinates)


    
powerplant_nodes = city_nodes[:4]

for item in powerplant_nodes:
    powerplant_coords.append(item.coordinates)

del city_nodes[:4]
del city_coords[:4]
    
coal_node_coord = dict(zip(coal_nodes, coal_coords))
city_node_coord = dict(zip(city_nodes, city_coords))
powerplant_node_coord = dict(zip(powerplant_nodes, powerplant_coords))

coal_pp_node_coords = {**coal_node_coord, **powerplant_node_coord}
pp_city_node_coords = {**powerplant_node_coord, **city_node_coord}
```

Setting up Distance Matrix  

```python

%pip install scikit-learn

import numpy as np
import sklearn

from sklearn.metrics import pairwise_distances


def getSquareMatrix(old_matrix):
    
    rows, cols = old_matrix.shape
    max_dim = max(rows, cols)
    
    square_mat = np.zeros((max_dim, max_dim))
    square_mat[:rows, :cols] = old_matrix
    
    return square_mat

def applyThreshold(matrix, threshold):
    matrix[matrix > threshold] = 0
    

# Two sets of binary data points
# Compute binary distance matrix between the two sets
dist_band = pairwise_distances(np.array(coal_coords), np.array(city_coords), metric='manhattan')


coal_powerplant_dist_band = pairwise_distances(np.array(coal_coords), np.array(powerplant_coords), metric='manhattan')
powerplant_city_dist_band = pairwise_distances(np.array(powerplant_coords), np.array(city_coords), metric='manhattan')

sq_coal_pp = getSquareMatrix(coal_powerplant_dist_band)
sq_pp_city = getSquareMatrix(powerplant_city_dist_band)

applyThreshold(sq_coal_pp, 55)
applyThreshold(sq_pp_city, 55)
```

```python

def getEdgeNames(edge_list):
    
    edge_names = {}
    
    for i in range(len(edge_list)):
        
        edge_names[edge_list[i]] = "edge" + str(i)
    
    return edge_names

edge_list_names = getEdgeNames([*all_edge_capacities.keys()])    
```

```python





def getConstraints(edge_capacities):
    
    constraints = [] #gets filled in
    
    
    
    #iterate over powerplant nodes
    for node in powerplant_nodes:
        
        #get the edges going INTO the powerplants(from the coal mines)
        in_edges = dist_graph.in_edges(node)
        
        #get the edges going OUT of the powerplants(into the cities)
        out_edges = dist_graph.out_edges(node)
        
        if len(in_edges) == 0 or len(out_edges) == 0: continue
        
        all_edges = list(in_edges) + list(out_edges)
        names = []
        
        constraint_values = []
        
        
        
        #iterate over the in_edges
        for edge in in_edges:
            
            names.append(edge_list_names[edge])
           
            
            #the in_edges need a constraint value of 1.0
            constraint_values.append(1 / len(in_edges))
            
        
        for edge in out_edges:
            
            names.append(edge_list_names[edge])
            
            #the out_edges need a constraint value of -1
            constraint_values.append(-1 / len(out_edges))
        
        #the constraints are a list of lists - names of edges associated with constraint values
        constraints.append([names, constraint_values])
            
    
    
    constraint_names = []
    
    for idx, val in enumerate(constraints):
        
        constraint_names.append(str(idx))
        
    constraint_senses = ['E'] * len(constraints) #should all be equality
    righthand_sides = [0] * len(constraints) #just zeroes
    
    
    return constraint_names, constraint_senses, constraints, righthand_sides
```

```python
#GitHub Gist Link: https://gist.github.com/stevenwalton/601645612161fb6ebf82dd7687de0060 
#Credit to Stephen Walton and Zayd H 

# Max Flow example from slides of UO 510: Multi-Agent Systems
# Class: https://classes.cs.uoregon.edu/19S/cis410mas/

# Flow moves to the right 
#      2
#   1-----3
# 4/ \     \3
# /   \     \
#s     \1    t
# \     \   /
# 5\     \ /2
#   2-----4 
#      4
# Nodes {s,1,2,3,4,t}
# Edges {s1(4), 13(2), 3t(3), 14(1), s2(5), 24(4), 4t(2)}
# Problem
# max_x x_{s1} + x_{s2}
#        (Conservation)             (Capacity)
# s.t. x_{s1} = x_{13} + x_{14}     x_{s1} <= 4
#      x_{s2} = x_{24}              x_{s2} <= 5
#      x_{13} = x_{3t}              x_{13} <= 2
#      x_{14} + x_{24} = x_{4t}     x_{14} <= 1
#                                   x_{24} <= 4
#                                   x_{3t} <= 3 
#                                   x_{4t} <= 2


try: 
    import cplex
except: 
    %pip install cplex

# Initialize

def getMaxFlowModel(edge_capacities):

    p = cplex.Cplex()
    p.set_problem_type(cplex.Cplex.problem_type.LP)
    p.objective.set_sense(p.objective.sense.maximize) # max problem

    #       0     1     2     3     4     5     6
    
    #v   = ["s1", "13", "3t", "14", "s2", "24", "4t"]  EQUIVALENT TO EDGE LIST
    
    
    obj_func = np.full(shape=len(edge_capacities), fill_value=1.0) #set all as ones
    low_bnd  = np.zeros(shape=len(edge_capacities))
    upr_bnd  = [*edge_capacities.values()] # Capacity constraints
    names = [*edge_list_names.values()]
    
    
    cnames, csenses, constraints, rhs = getConstraints(edge_capacities)
    print(constraints)
    
    p.variables.add(obj = obj_func.tolist(), lb=low_bnd.tolist(), ub=upr_bnd, names=names)
    
    
    p.linear_constraints.add(lin_expr=constraints,
                            senses = csenses,
                            rhs = rhs,
                            names = cnames)
    
    return p
```
